```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
speeds = pd.read_csv("speeds.csv")
speeds_clip = pd.read_csv("speeds_acc.csv")
```

```python
speeds = speeds['Speed']
speeds_clip = speeds_clip['Speed']
```

```python
speeds = np.array(speeds)
speeds_clip = np.array(speeds_clip)
```

```python
speeds.shape, speeds_clip.shape
```

```python
plt.figure(figsize=(10, 6))
plt.plot(speeds, label='Speeds')
plt.plot(speeds_clip, label='Speeds Clip')
plt.xlabel('Index')
plt.ylabel('Speed')
plt.title('Speeds vs Speeds Clip')
plt.legend()
plt.show()
```
