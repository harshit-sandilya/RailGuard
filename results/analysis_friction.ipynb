```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
error = pd.read_csv('friction_error.csv')
pid = pd.read_csv('friction_pid.csv')
```

```python
error.head()
```

```python
error = np.array(error['FrictionForceMagnitude'])
pid = np.array(pid['FrictionForceMagnitude'])
```

```python
error.shape, pid.shape
```

```python
error = error[0:500]
pid = pid[0:500]
```

```python
error = error/np.min(error)
pid = pid/np.min(pid)
```

```python
plt.figure(figsize=(20, 12))
plt.plot(error, label='Error')
plt.plot(pid, label='PID')
plt.xlabel('Index')
plt.ylabel('Friction Force Magnitude')
plt.title('Error vs PID')
plt.legend()
plt.show()
```
